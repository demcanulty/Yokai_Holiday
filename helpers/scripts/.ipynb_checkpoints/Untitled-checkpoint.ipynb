```python
import matplotlib 
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

```python
inptr = np.int32()
outptr = np.int32()
target_buffer_length = 124915

# inptr_arr = np.array(0, dtype='int32')
# outptr_arr = np.array(0, dtype='int32')

inptr_arr  = []
outptr_arr = []
```

```python
for i in range(target_buffer_length * 2):
    
    inptr_arr.append(inptr)
    outptr_arr.append(outptr)
    
    inptr +=1
    outptr -=1
    
    if(outptr < 0):
        outptr = target_buffer_length - 1
        
    if(inptr >= target_buffer_length):
        inptr = 1
```

```python
inptr_np_arr = np.array(inptr_arr, dtype='int32')
outptr_np_arr = np.array(outptr_arr, dtype='int32')
```

```python
plt.plot(inptr_np_arr)
plt.plot(outptr_np_arr)

plt.plot(inptr_np_arr - outptr_np_arr)
```

```python
diff_arr = []

for i in range(len(inptr_arr)):
    this_inptr = inptr_arr[i]
    this_outptr = outptr_arr[i]
    
    if(this_inptr > this_outptr):
        diff = this_inptr - this_outptr
    else:
        diff = this_outptr - this_inptr
        
    diff_arr.append(diff)
```

```python
plt.plot(inptr_np_arr)
plt.plot(outptr_np_arr)
plt.plot(diff_arr)
```

```python
plt.plot(inptr_np_arr)
plt.plot(outptr_np_arr)

plt.plot(abs(inptr_np_arr - outptr_np_arr))
```

```python
new_diff = []

for i in range(len(inptr_arr)):
    this_inptr = inptr_arr[i]
    this_outptr = outptr_arr[i]
    
    if(this_inptr > this_outptr):
        top_distance = target_buffer_length - this_inptr
        bot_distance  = this_outptr
        diff = top_distance + bot_distance
    else:
        diff = this_outptr - this_inptr
        
    new_diff.append(diff)
```

```python
plt.plot(inptr_np_arr)
plt.plot(outptr_np_arr)
plt.plot(new_diff)
```
